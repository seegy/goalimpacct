##  can we check favourite and hate teams of referees by cards?
### Whats the biggest average difference between all referees and one specific referee by team? (More cards)

```python
targetSeasions = ['2019', '2018', '2017', '2016', '2015', \
                  '2014', '2013', '2012', '2011', '2010']

blLastXYearsDF = matchDF[  (matchDF['tournament'] == '1. Bundesliga') 
                         & (matchDF['saison'].isin(targetSeasions))]


bllXyguestDF = blLastXYearsDF[['matchid', 'guestteam', 'refereeid']].rename(columns={"guestteam": "teamid"})
bllXyhomeDF = blLastXYearsDF[['matchid', 'hometeam', 'refereeid']].rename(columns={"hometeam": "teamid"})

bllXyTeamDF = pd.concat([bllXyguestDF, bllXyhomeDF])


cardsAndMatchTeamsDF = pd.merge(bllXyTeamDF, cardDF, on=['matchid','teamid'])

cardCountPerMatch = pd.DataFrame({
    'count' : cardsAndMatchTeamsDF.groupby(['refereeid', 'teamid', 'matchid']).size()}).reset_index()

avgCardsPerTeamAndReferee = pd.DataFrame({
    'avgReferee' : cardCountPerMatch.groupby(['refereeid', 'teamid'])['count'].mean()}).reset_index()

avgCardsPerTeam = pd.DataFrame({
    'avgTeam' : avgCardsPerTeamAndReferee.groupby(['teamid'])['avgReferee'].mean()}).reset_index()

refereeMatchCount = pd.DataFrame({ 'matchCount' : bllXyTeamDF.groupby(['refereeid', 'teamid']).size()}).reset_index()

diffCards = pd.merge(avgCardsPerTeam, avgCardsPerTeamAndReferee, on="teamid")
diffCards = pd.merge(diffCards, refereeMatchCount)
diffCards = diffCards[diffCards['matchCount'] > 2]
diffCards['diff'] = diffCards['avgReferee'] - diffCards['avgTeam']
diffCards['diff_max'] = diffCards.groupby(['refereeid'])['diff'].transform(max)
diffCards['diff_min'] = diffCards.groupby(['refereeid'])['diff'].transform(min)

refereesMostCards = diffCards[diffCards['diff_max'] == diffCards['diff']][['refereeid', 'teamid', 'diff', 'matchCount']] 

refereesMostCards = pd.merge(pd.merge(refereesMostCards, teamDF, on='teamid'), \
        refereeDF, \
        on='refereeid').sort_values(by=['diff'], ascending=False)

refereesMostCards
```

### (lesser cards)

```python
refereesLeastCards = diffCards[diffCards['diff_min'] == diffCards['diff']][['refereeid', 'teamid', 'diff', 'matchCount']] 

refereesLeastCards = pd.merge(pd.merge(refereesLeastCards, teamDF, on='teamid'), \
        refereeDF, \
        on='refereeid').sort_values(by=['diff'], ascending=True)

refereesLeastCards
```

## So what are the favourite and hate team of each referee?

```python
refereesFavTeam = refereesLeastCards[['refereeid', 'refereename','teamname']] \
                        .rename(columns={'teamname': 'Favourite Team'})
refereesHateteam = refereesMostCards[['refereeid', 'refereename','teamname']] \
                        .rename(columns={'teamname': 'Hate Team'})

refereesFavHateTeam = pd.merge(refereesFavTeam, refereesHateteam, on=['refereeid', 'refereename']).drop(columns=['refereeid'])

refereesFavHateTeam = refereesFavHateTeam[refereesFavHateTeam['Hate Team'] != refereesFavHateTeam['Favourite Team']] \
                        .rename(columns={'refereename': 'Referee'})

refereesFavHateTeam
```

## which player has the most playtime (per tournament and saison)

```python
matchAndtimes = pd.merge(matchDF[(pd.to_numeric(matchDF['saison'], errors='coerce') >= 2015)], \
                         playertimeDF)
matchAndtimes
```

```python
playerTimesPerTS = matchAndtimes.groupby(['tournament', 'saison', 'playerid'])['playtime'].sum().reset_index()
playerTimesPerTS['playtime_max'] = playerTimesPerTS.groupby(['tournament', 'saison'])['playtime'].transform(max)
playerTimesPerTS = playerTimesPerTS[ playerTimesPerTS['playtime'] == playerTimesPerTS['playtime_max'] ] \
                        .drop(columns=['playtime'])
#playerTimesPerTS
```

## How often relates which position to the player with the most playtime per tournament and saison? 

```python
playerDetails = pd.merge(playerTimesPerTS, profileDF) \
                                     .drop(columns=['birthdate', 'nationality', 'playerid'])

positionTimesMaxPT  =  pd.DataFrame({'count' : playerDetails
                                     .groupby(['position']).size()}).reset_index()

#positionTimesMaxPT
```

```python
fig1, ax1 = plt.subplots()
ax1.pie(positionTimesMaxPT['count'], labels=positionTimesMaxPT['position'], autopct='%1.1f%%',
        shadow=True, startangle=90, radius=2.5)

1 #avoid plt outpout
```
